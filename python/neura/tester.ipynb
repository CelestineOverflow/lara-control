```python
from neurapy.robot import Robot
import time

robot = Robot()
robot.turn_on_jog(
    jog_velocity=[0, 0, 1, 0, 0, 0], 
    jog_type='Cartesian'
)
robot.jog(set_jogging_external_flag = 1)
i = 0
start = time.time()
while(i < 1000):
    robot.jog(set_jogging_external_flag = 1)
    i += 1
    elapsed_time = (time.time() - start) * 1000  # Convert to milliseconds
    print(f"Iteration {i}: {elapsed_time:.2f} ms")
    start = time.time()
robot.turn_off_jog()
```
